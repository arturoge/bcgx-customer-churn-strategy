## Data visualization

If you're working in Python, two of the most popular packages for visualization are `matplotlib` and `seaborn`. We highly recommend you use these, or at least be familiar with them because they are ubiquitous!

Below are some functions that you can use to get started with visualizations.

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:

        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

Thhe first function `plot_stacked_bars` is used to plot a stacked bar chart. An example of how you could use this is shown below:

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_="lower right")
```

The second function `annotate_bars` is used by the first function, but the third function `plot_distribution` helps you to plot the distribution of a numeric column. An example of how it can be used is given below:

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
```

## Analyzing Sales Channels

```python
# --- Step 1: Analyze Churn by Sales Channel ---

# Group data by sales channel and churn status
channel = client_df[['id', 'channel_sales', 'churn']]
channel = channel.groupby([channel['channel_sales'], channel['churn']])['id'].count().unstack(level=1).fillna(0)

# Calculate the percentage of churn for each channel
channel_churn = (channel.div(channel.sum(axis=1), axis=0) * 100).sort_values(by=[1], ascending=False)

# Plot the results using the function defined earlier
plot_stacked_bars(channel_churn, 'Sales Channel Attrition', rot_=30)
```

## Analyzing Contract Antiquity

```python
# --- Step 2: Analyze Churn by Antiquity (Years) ---

# Group data by the number of years (antiquity) and churn status
consumption = client_df[['id', 'num_years_antig', 'churn']]
dates = consumption.groupby([consumption['num_years_antig'], consumption['churn']])['id'].count().unstack(level=1).fillna(0)

# Calculate the percentage of churn for each year
dates_churn = (dates.div(dates.sum(axis=1), axis=0) * 100).sort_values(by=[1], ascending=False)

# Plot the results
plot_stacked_bars(dates_churn, 'Churn by Antiquity (Years)', rot_=30)
```

## Price Sensitivity Analysis

```python
# --- Step 3: Price Sensitivity Analysis ---

# Hypothesis: Is churn driven by price sensitivity?
# We will check the distribution of energy and power prices.

# Create a figure with 2 subplots (side by side)
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(18, 5))

# Plot 1: Distribution of Energy Price (Off-Peak)
# This is the variable cost per kWh
sns.histplot(price_df, x='price_off_peak_var', ax=axs[0], bins=50, kde=True, color='skyblue')
axs[0].set_title('Distribution of Energy Price (Off-Peak)')

# Plot 2: Distribution of Power Price (Fixed)
# This is the fixed monthly cost
sns.histplot(price_df, x='price_off_peak_fix', ax=axs[1], bins=50, kde=True, color='orange')
axs[1].set_title('Distribution of Power Price (Fixed)')

plt.tight_layout()
plt.show()
```

## Key Findings & Recommendations

## Key Findings

After performing the Exploratory Data Analysis (EDA), we have identified several key drivers for customer churn:

1.  **Churn Rate:** The overall churn rate is approximately **9.7%**. This is the baseline we need to beat with our predictive model.
2.  **Sales Channels:** We identified that the sales channel encoded as `foosdfpfkusacimwkcsosbicdxkicaua` has a significantly higher churn rate (**~12%**) compared to the average. This suggests a potential issue with the sales strategy or customer quality in that specific channel.
3.  **Contract Antiquity:** There is a noticeable spike in churn at **4 years** of tenure. This likely coincides with the end of long-term contracts. PowerCo should consider a proactive retention campaign for customers approaching their 4th anniversary.
4.  **Price Sensitivity:**
    * Prices for energy and power are highly concentrated.
    * Visual inspection alone does not show a linear relationship between higher prices and churn.
    * **Next Step:** We need to calculate specific features (like price changes over time) in the Feature Engineering phase to better capture price sensitivity.
